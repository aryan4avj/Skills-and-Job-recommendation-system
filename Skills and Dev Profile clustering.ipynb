```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sys
print(sys.path)

from kmodes.kmodes import KModes
```

```python
# SKills Clustering

df = pd.read_csv("./data/skills and dev profile clustering/Cleaned dataset.csv")
df = df.fillna(0)
df = df.drop(df.index[5000:])
# df.head()

km = KModes(n_clusters=30, init='Huang', n_init=11, verbose=1)
# fit the clusters to the skills dataframe
clusters = km.fit_predict(df)
# get an array of cluster modes
kmodes = km.cluster_centroids_
shape = kmodes.shape
# For each cluster mode (a vector of "1" and "0")
for i in range(shape[0]):
    if sum(kmodes[i,:]) == 0:
        print("\ncluster " + str(i) + ": ")
        print("no-skills cluster")
    else:
        print("\ncluster " + str(i) + ": ")
        cent = kmodes[i,:]
        for j in df.columns[np.nonzero(cent)]:
            print(j)
```
